```python
import numpy as np

def run_simulation(disparity_bias=1.0, label="Baseline"):
    n_sites = 100000
    # 1. Generate 3 vectors of random numbers (Uniform 0 to 1)
    disparity = np.random.uniform(0, 1, n_sites) * disparity_bias
    motion = np.random.uniform(0, 1, n_sites)
    color = np.random.uniform(0, 1, n_sites)
    
    # 4. Extracting Properties using the sqrt(2) rule
    sqrt2 = np.sqrt(2)
    selective_d, selective_m, selective_c, overlap_count = 0, 0, 0, 0
    
    for i in range(n_sites):
        vals = np.array([disparity[i], motion[i], color[i]])
        labels = ['Disparity', 'Motion', 'Color']
        
        idx_sorted = np.argsort(vals) # [min, mid, max]
        v_max = vals[idx_sorted[2]]
        v_mid = vals[idx_sorted[1]]
        
        # A site is "Selective" if Max > sqrt(2) * second highest
        if v_max > (sqrt2 * v_mid):
            winner = labels[idx_sorted[2]]
            if winner == 'Disparity': selective_d += 1
            elif winner == 'Motion': selective_m += 1
            else: selective_c += 1
        # A site is "Overlap" if the ratio is smaller than sqrt(2)
        elif (v_max / (v_mid + 1e-9)) < sqrt2:
            overlap_count += 1
            
    print(f"--- {label} ---")
    print(f"Total Sites Passing Threshold: {n_sites}")
    print(f"Selective Disparity: {selective_d} ({selective_d/n_sites*100:.1f}%)")
    print(f"Selective Motion:    {selective_m} ({selective_m/n_sites*100:.1f}%)")
    print(f"Selective Color:     {selective_c} ({selective_c/n_sites*100:.1f}%)")
    print(f"Overlap Sites:       {overlap_count} ({overlap_count/n_sites*100:.1f}%)")
    print("-" * 35)

# Case 1: All features have equal random noise
run_simulation(disparity_bias=1.0, label="Baseline (Equal Noise)")

# Case 2: Disparity input is significantly lower/absent
run_simulation(disparity_bias=0.5, label="Low Disparity Input")
```
